```r
# ***********************************************************************************************
# Titel: Exempel p diagramtyper i R - HEXBIN PLOT 
# Beskrivning: Illustrera exempel med diagramtyper i R
# Skapad av: Frank Weideskog(frank.weideskog@scb.se), ID: SCBFRWE
# 2024-06-17 (SCBFRWE):En frsta version 
#*************************************************************************************************
# Rensar minnet i R
rm(list=ls())				
# ************************************************************************************************
#Installerar/aktiverar relevanta R-paket (R-packages):
suppressWarnings({
  suppressMessages({
    install.packages("hexbin")
    install.packages("RColorBrewer")
    library(hexbin)
    library(RColorBrewer)
    library(ggplot2)
    library(ggplot2)      
  }) 

# ************************************************************************************************

# HEXBIN CHART

suppressPackageStartupMessages(library(hexbin))
suppressPackageStartupMessages(library(RColorBrewer))
suppressPackageStartupMessages(library(ggplot2))

text<-'Hexbin r en typ av diagram som anvnds fr att visualisera densitet och frdelning av datapunkter tvdimensionellt. Istllet fr att anvnda traditionella punkter eller linjer, anvnder hexbin plot hexagonala bin fr att representera frdelningen av datapunkter. Varje hexagon representerar ett omrde av datarummet och frgen eller intensiteten av hexagonen indikerar antalet datapunkter som finns inom omrdet. Hexbin plots r srskilt anvndbara vid en stor mngd datapunkter som behver visualisers, och gr det lttare att identifiera mnster och kluster. I detta exempel genererar vi tv vektorer med slumpmssiga datapunkter fr x- och y-variablerna. Sedan skapar vi en hexbin plot med hjlp av hexbin()-funktionen frn hexbin-paketet och anger antalet binar (xbins) fr att kontrollera upplsningen av hexagonerna. Slutligen anvnder vi plot()-funktionen fr att rita hexbin plot med den skapade hexbin-objektet och specificerar huvudtiteln och axelbeskrivningarna. Resultatet r en hexbin plot som visualiserar frdelningen av datapunkterna i det tvdimensionella datarummet med hexagonala bin.'
text


# Create data
x <- rnorm(mean=1.5, 5000)
y <- rnorm(mean=1.6, 5000)

bin<-hexbin(x, y, xbins=40)
my_colors=colorRampPalette(rev(brewer.pal(11,'Spectral')))
p <- plot(bin, main="Hexbin chart" , colramp=my_colors , legend=F ) 

p


# HEXBIN PLOT

# data
sample_data <- data.frame(x_values = 1:100 + rnorm(100,sd=20), y_values = 1:100 + rnorm(100,sd=27)) 

#plot
p <- ggplot(sample_data, aes(x_values, y_values)) +
  stat_density2d(geom="tile", aes(fill = ..density..), contour = FALSE) + 
  geom_point(colour = "white") +
  labs(title = "Hexbin plot")

p

 }) 
```
