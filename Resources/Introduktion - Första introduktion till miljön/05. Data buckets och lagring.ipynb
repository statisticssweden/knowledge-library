# 05. Data buckets och lagring

## 1. Vad är en bucket?

"Bucket" är en term som oftast används inom datalagring och är vanlig inom molnbaserad lagring. En bucket är i grunden en behållare som används för att lagra data, som filer, dokument eller objekt. Det fungerar som en toppnivåmapp eller behållare där du kan organisera och lagra dina data.Här är några viktiga aspekter att förstå om buckets.

I Balsam där vi använder MinIo har vi ställt in så att data lagras i den egna infrastrukturen. 

### Om buckets i MinIO:
* Dataorganisation: Ett MinIO-bucket fungerar som en behållare för att organisera och lagra objekt (data) på MinIO-servern. Man kan se det som en mapp eller katalog där lagring av dina filer och data jan göras.
* Unik namngivning: Namnet på ett MinIO-bucket måste vara unikt inom MinIO-serverns namnrymd. Med andra ord, två olika användare eller applikationer kan inte skapa två buckets med samma namn.
* Åtkomstkontroll: Du kan använda MinIO:s åtkomstkontrollsfunktioner för att styra vem som har rättigheter att läsa, skriva och ta bort objekt från ett specifikt bucket. Det ger dig möjlighet att säkerställa att ditt data är säkert och skyddat.
* Skalbarhet: MinIO är designat för att vara skalbart, och du kan skapa och hantera ett stort antal buckets beroende på dina behov.
* Dataklassificering: Eftersom buckets är en grundläggande organisationsenhet, kan du använda dem för att kategorisera och separera data. Till exempel kan du ha ett bucket för användardata, ett annat för loggfiler och ett tredje för säkerhetskopiering.
* I huvudsak är en bucket i MinIO en viktig komponent som hjälper dig att organisera och hantera dataeffektivt på din MinIO-server. Det är också användbart för att kontrollera åtkomsträttigheter och säkerställa att dina data är strukturerat och skyddat.

## 2. Vem kan läsa och skriva till  din MinIO bucket?

I MinIO kan du styra åtkomsträttigheterna för dina buckets med hjälp av användar- och policyhantering. Åtkomsträttigheterna för dina 
MinIO buckets kan konfigureras på följande sätt:

* Användaråtkomst: Du kan skapa användarkonton och tilldela dem specifika åtkomsträttigheter till dina buckets. Du kan ange om en användare har rätt att läsa, skriva, uppdatera eller ta bort objekt från ett visst bucket. Detta ger dig fin kontroll över vem som kan göra vad med dina data.
* Policybaserad åtkomst: Utöver användarkonton kan du skapa policyer som definierar regler för åtkomst till dina buckets. Med hjälp av policyer kan du definiera åtkomstregler som kan tillämpas på ett eller flera buckets samtidigt. Du kan använda dessa policyer för att begränsa eller tillåta åtkomst baserat på kriterier som användargrupper eller IP-adresser.
* Delade åtkomstlänkar: MinIO stöder även möjligheten att generera tidsbegränsade delade åtkomstlänkar (presigned URLs). Med dessa länkar kan du tillfälligt ge någon åtkomst till specifika objekt i dina buckets utan att behöva dela dina användaruppgifter. Länkarna kan ha en begränsad giltighetstid och begränsningar för åtkomst, vilket ökar säkerheten.

Sammanfattningsvis kan endast de användare eller enheter som har rätt behörighet och godkända åtkomstreferenser (till exempel användarnamn och lösenord) läsa och skriva till dina MinIO buckets. Du kan kontrollera dessa åtkomsträttigheter genom att konfigurera användare, policyer och delade åtkomstlänkar enligt dina behov för att säkerställa att din data är skyddad och endast nås av auktoriserade parter.

## 3.  Hur använder vi buckets i Balsam?
Vi använder buckets för att lagra data kopplat till bearbetning och analys. Detta för att få till en säker hantering av data samtidigt som den gör spårbarhet och reproducerbarhet möjliga.

## 4. Hur får jag in data i bucket-miljön i MinIo?
För att lägga till data (objekt) i en MinIO-bucket, kan du använda några olika metoder. Här är några sätt att lägga till data i en MinIO-bucket:

### 1. MinIO Web UI (användargränssnitt):
* Gå till MinIO Web UI genom att öppna din webbläsare och navigera till MinIO-serverns webbgränssnitt (vanligtvis tillgängligt på 'http://**minio-server**:9000' om du inte har konfigurerat en annan port).
* Logga in med dina användaruppgifter.
* I webbgränssnittet kan du använda alternativet "Upload" eller "Drag and Drop" för att ladda upp filer eller objekt till ditt önskade bucket.


### 2. MinIO Command Line Interface (mc):
* Använd MinIO:s kommandoradsverktyg, mc, för att lägga till objekt i en bucket. Du kan använda kommandon som mc cp för att kopiera filer till en bucket.
* Exempel: 'mc cp /lokalt/fil.pdf myminio/bucket/fil.pdf'


### 3. Programmatisk åtkomst via SDK eller API:
* Om du är en utvecklare och vill automatisera dataöverföringar till MinIO kan du använda MinIO SDK:er eller Amazon S3 SDK:er (MinIO är kompatibelt med S3) för att skriva skript eller applikationer som lägger till objekt i en bucket.
* Exempel i Python med Boto3 SDK (som används för att interagera med Amazon S3, men fungerar också med MinIO):
    * Python kod:
        * `import boto3`
        * `#Skapa en S3-klient`
        * `s3 = boto3.client('s3', endpoint_url='http://minio-server:9000', aws_access_key_id='your-access-key', aws_secret_access_key='your-secret-key') `
        * `#Ladda upp en fil till en bucket`
        * `s3.upload_file('lokalt/fil.pdf', 'mybucket', 'fil.pdf')`

Oavsett vilken metod du väljer, måste du se till att du har rätt åtkomst och behörighet för att lägga till objekt i det önskade bucketet. Detta kan konfigureras genom användar- och policyhantering, som tidigare nämnts.

## 5. Hur sparar jag data till en bucket i MinIo?
* För att spara data (objekt) i en MinIO-bucket kan du använda MinIO:s kommandoradsverktyg (mc) eller programmatisk åtkomst via SDK eller API. Här är steg-för-steg-instruktioner för att spara data i en MinIO-bucket med hjälp av några olika metoder:

### Metod 1: Använd MinIO Command Line Interface (mc)
* Öppna ett kommandotolkfönster eller terminal.
* Använd mc-kommandot för att kopiera en lokal fil eller ett objekt till din MinIO-bucket. 
    * Syntaxen är: `mc cp /lokalt/fil.pdf myminio/bucket/fil.pdf`
    * `/lokalt/fil.pdf` är sökvägen till den lokala filen som du vill kopiera.
    * myminio är namnet på din MinIO-server.
    * bucket är namnet på det målbucket där du vill spara filen.
    * fil.pdf är det namn som objektet kommer att ha i bucketen.

### Metod 2: Använd MinIO Python SDK (Boto3)
* Om du vill spara data i en MinIO-bucket programmatiskt, kan du använda ett SDK som Boto3 (ursprungligen för Amazon S3, men kompatibelt med MinIO). Här är ett Python-exempel med Boto3:

**Skapa en S3-klient**
`s3 = boto3.client('s3', endpoint_url='http://minio-server:9000', aws_access_key_id='your-access-key', aws_secret_access_key='your-secret-key')`

**Ladda upp en fil till en bucket**
`s3.upload_file('lokalt/fil.pdf', 'mybucket', 'fil.pdf')`

### Metod 3: Använd MinIO JavaScript SDK
* Om du utvecklar en webbapplikation och vill använda JavaScript, kan du använda MinIO:s JavaScript SDK. 

Oavsett vilken metod du väljer måste du se till att du har rätt åtkomst och behörigheter för att spara data i det önskade bucketet. Detta kan konfigureras genom användar- och policyhantering, som tidigare nämnts.

## 7. Hur tittar jag på data som ligger i en MinIo bucket?
* För att titta på data som ligger i en MinIO-bucket kan du använda olika metode., Här är några sätt att granska data som finns i en MinIO-bucket:

### Metod 1: Använd MinIO Web UI (användargränssnitt)
* MinIO tillhandahåller ett webbgränssnitt som gör det enkelt att bläddra igenom och titta på data i dina buckets. Följ dessa steg:
* Öppna din webbläsare och gå till MinIO Web UI genom att navigera till din MinIO-servers webbadress (vanligtvis tillgänglig på 'http://**minio-server**>:9000' om du inte har konfigurerat en annan port).
    * Logga in med dina användaruppgifter.
    * I webbgränssnittet kan du navigera till det bucket där du vill titta på data. Klicka på bucketnamnet för att öppna det.
    * Du kan se en lista över objekt som finns i bucketen. Klicka på ett objekt för att visa eller ladda ner det.

### Metod 2: Använd MinIO Command Line Interface (mc)
* MinIO:s kommandoradsverktyg (mc) kan användas för att lista objekt i en bucket och ladda ner dem från kommandotolken. Använd kommandon som mc ls och mc cp för att lista och hämta objekt i bucketen.
    * Exempel:
        * För att lista objekt i en bucket:
            * `mc ls myminio/bucket`
        * För att ladda ner ett objekt från en bucket:
            * `mc cp myminio/bucket/fil.pdf /lokalt/mål`

### Metod 3: Använd programmatisk åtkomst via SDK eller API
* Om du är en utvecklare kan du använda MinIO SDK:er eller Amazon S3 SDK:er för att lista och hämta objekt från en MinIO-bucket. Här är ett exempel i Python med Boto3 SDK:

`import boto3`

**Skapa en S3-klient**
`s3 = boto3.client('s3', endpoint_url='http://**minio-server**:9000', aws_access_key_id='your-access-key', aws_secret_access_key='your-secret-key')`

**Lista objekt i en bucket**
`objects = s3.list_objects(Bucket='mybucket')`
`for obj in objects.get('Contents', []):`
    `print('Objektnamn:', obj['Key'])`

**Hämta ett objekt från en bucket**
`s3.download_file('mybucket', 'fil.pdf', 'lokalt/mål/fil.pdf')`

Oavsett vilken metod du väljer, måste du ha rätt åtkomst och behörighet för att visa och hämta data från det aktuella bucketet. Åtkomsträttigheter kan konfigureras genom användar- och policyhantering enligt dina krav och säkerhetsregler.
