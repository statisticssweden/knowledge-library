# Jupyter Notebook och tillhörande delar

## 1. Vad är en Jupyter Notebook?
Jupyter Notebook (JN) är ett webbaserat verktyg för öppen kodbehandling där man interaktivt kan skriva och köra R- och Pythonkod och generera resultat i en sorts anteckningsbok. Namnet Jupyter är en sammandragning av programspråken Julia, Python och R.

För att kunna programmera, behöver man någonting att skriva koden i. Koden kan skrivas var som helst, men det underlättar om man har någon form av IDE – Integrated Development Kit (utvecklingsmiljö) att skriva koden i. I en IDE, som exempelvis JN, kan man bland annat få hjälp med att hitta fel i koden och få förslag på en förbättrad kodning.

JN-filen innehåller en kombination av programmeringskod, Markdown och utdata genererade från koden, och kan beskrivas som en interaktiv miljö där en utvecklare kan använda för att exekvera kodblock och lägga till områden för dokumentation för att förklara själva koden.

JN är inget fristående analysverktyg utan mer ett redskap för att arbeta med programkod som analysverktyg eller för att presentera resultaten samt bakomliggande arbetsprocesser.

### Sammanfattande beskrivning av JN:
- En utvecklingsmiljö där man kan skriva och köra sin kod (R/Python)
- Ger möjlighet till presentation och förklaringar tillsammans med koden, vilket ger en pedagogisk och tydligare bild av vad koden gör
- Kan köra kod direkt i webbläsaren oberoende av andra program.

## 2. Vilka funktioner finns i Jupyter Notebooks?
JN-filen kan användas för att både exekvera och dokumentera kod. Det förutsätter ingen komplicerad installation, har en simpel användarmiljö och går smidigt att dela. Man kan lätt testa delar av koden och redigera separata avsnitt av dokumentet.

JN har stora fördelar för undervisning och används ofta vid explorativ dataanalys, demonstrationer av olika slag samt vid framtagning av undervisningsmaterial. Det går bra att skapa, hantera och ändra Notebooks lokalt på egen dator såväl som över nätet.

### Fördelar:
- Lättillgängligt sätt att köra programkod.
- Kan exekvera R- och Pythonkod direkt i webbläsaren.
- Möjlighet till presentation och förklaringar tillsammans med koden.
- Autosparar allteftersom man arbetar i den.

## 3. Vilket filformat har Jupyter Notebooks?
Filformatet i JN är .ipynb (IPython Notebook), men kan också sparas i HTML- eller PDF-format. Filformatet tillåter kombinationer av avsnitt med text, bilder och annan media med avsnitt som innehåller programkod och resultat i samma fil.

## 4. Vad är JupyterLab och JupyterHub?
### JupyterLab
JupyterLab är en vidareutveckling och framtida ersättare för gränssnittet till JN och skriver samma filformat som JN.

### JupyterHub
JupyterHub är en serverprogramvara som gör JN och/eller JupyterLab tillgängliga för användare över Internet.

## 5. Vad är ett Markdowndokument?
Ett Markdowndokument kan liknas vid en rapportgenerator där dokumentet skrivs i Markdown och innehåller körbara kodbitar.

## 6. Hur länkas andra media till anteckningsboken?
En JN kan ha material inlänkat från andra platser, som sökvägar till filer i filsystemet, eller länkar till filer på internet, Youtube-klipp, etc. Media och data kan också bäddas in som en del av både Markdown och kod.

## 7. Hur exporteras en anteckningsbok?
En JN kan exporteras till olika presentationsformat som AsciiDoc, HTML, LaTeX, Markdown, PDF, reStructuredText, Reveal.js slides via File-menyn.

## 8. Vad är Nbviewer?
Nbviewer är en webbapplikation som renderar anteckningsboken som en statisk HTML-webbsida och ger en stabil länk till den sidan.

## 9. Vad är Mybinder.org?
Mybinder.org är en webbtjänst som låter dig öppna anteckningsböcker i en körbar miljö, vilket gör din kod omedelbart reproducerbar av vem som helst, var som helst.

## 10. Hur lägger jag in min kod och skapar en JN?
JN är en anteckningsbok som integrerar kod och utdata i ett enda dokument, och kombinerar visualiseringar, berättande text, matematiska ekvationer samt media.

### Skapa en notebook i R:
1. Klicka på R-ikonen för att skapa en Jupyter Notebook (JN) i R.
2. Döp om JN genom att i vänstra listen söka upp den nya JN och välja ”rename”.
3. Skriv en enkel kod i den första cellen och exekvera/köra koden.
4. Lägg till ytterligare en cell och exekvera även den.
